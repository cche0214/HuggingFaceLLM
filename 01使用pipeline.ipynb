```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
classifier("I've been waiting for a HuggingFace course my whole life.")
```

```python
# pipeline是listin和listout
classifier([
    "This course is okey.",
    "I hate this.",
    "Absolutely amazing experience."
])
```

```python
# 换一个中文模型下载下来看一看
classifierZH = pipeline("sentiment-analysis", model="uer/roberta-base-finetuned-dianping-chinese")
```

```python
classifierZH([
    "操你妈",
    "我爱你",
    "我想操你"
])
```

```python
summarizer = pipeline("summarization")
```

```python
text = """
Hugging Face provides an ecosystem of libraries and tools for building, training,
and deploying machine learning models. Transformers is a popular library offering
pretrained models for NLP and beyond, enabling quick experimentation and production use.
"""
summarizer(text, max_length=30, min_length=20, do_sample=False)
```

```python
import requests

img_url = "https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/coco_sample.png"
img_path = "sample.png"
with open(img_path, "wb") as f:
  f.write(requests.get(img_url).content)
```

```python
img_path
```

```python
i2t = pipeline("image-to-text")
out = i2t(img_path)
print(out)
```

```python
ic = pipeline("image-classification")
```

```python
out = ic(img_path)
print(out)
```

```python
det = pipeline("object-detection")
```

```python
out = det(img_path)
print(out)
```

```python
# 零样本分类 不需要在自己的数据上对模型进行微调就可以使用 可以直接为自己的标签返回概率分数
classifierZero = pipeline("zero-shot-classification")
```

```python
classifierZero(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"]
)
```

```python
classifierZero(
    [
        "I want to make money.",
        "Transformers models is very usefull.",
        "driving after drinking is nor allow."
    ],
    candidate_labels=["education", "politics", "business"]
)
```
